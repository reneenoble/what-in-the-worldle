```python
import matplotlib.pyplot as plt

# data = {3: 18, 5: 22, 4: 29, None: 5, 6: 9, 2: 2}

dist = {3: 18, 5: 22, 4: 29, 6: 9, 2: 2, 1:0}

data = sorted(dist.items(), reverse=False)
x = [i[0] for i in data]
y = [i[1] for i in data]

fig, ax = plt.subplots()
hbars = ax.barh(x, y, color="#6aaa64", label=y)
ax.bar_label(hbars, y)
ax.invert_yaxis()
plt.show()
```
